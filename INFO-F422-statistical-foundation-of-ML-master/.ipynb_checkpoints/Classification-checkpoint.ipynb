```python
class Plotter():
  def __init__(self, width, height ,col=11, row=11):
      self.width, self.height = width, height
      figure = plt.figure(figsize=(col, row))

  def setup(self, X, h=0.2):

      self.x_min, self.x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
      self.y_min, self.y_max = X[:, 1].min() - .5, X[:, 1].max() + .5

      self.xx, self.yy = np.meshgrid(np.arange(self.x_min, self.x_max, h),
                          np.arange(self.y_min, self.y_max, h))

      # just plot the dataset first
      self.cm = plt.cm.RdBu
      self.cm_bright = ListedColormap(['#FF0000', '#0000FF'])
      self.alpha = 0.3
 


  def plot(self):

    plt.tight_layout() 
    plt.show()

  def plot_result(self, X_train, y_train, X_test, y_test, ds_cnt, clf, i, title, inSample=False):
      xx, yy = self.xx, self.yy
      width, height = self.width, self.height
      cm, cm_bright, alpha = self.cm, self.cm_bright, self.alpha
      cmcontourf = plt.cm.RdBu
      cmcontourf.set_under('#FF0000')
      cmcontourf.set_over('#0000FF')
      # Plot the decision boundary. For that, we will assign a color to each
      # point in the mesh [x_min, x_max]x[y_min, y_max].

      if hasattr(clf, "decision_function"):
          Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
      else:
          Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]

      # Put the result into a color plot
      Z = Z.reshape(xx.shape)

      ax = plt.subplot(width, height + 1, i)

      ax.contourf(xx, yy, Z, cmap=cmcontourf,  alpha=0.5, levels=[-100,0,100])

      # Plot the training points
      ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train,cmap=cm_bright, alpha=alpha,
                edgecolors='k')
 

      ax.set_xlim(xx.min(), xx.max())
      ax.set_ylim(yy.min(), yy.max())
      ax.set_xticks(())
      ax.set_yticks(())
        
      # Compute the score
      score = clf.score(X_test, y_test)    

      if ds_cnt == 0:
          ax.set_title(title)
      if inSample == True:
            ax.text(xx.max() - .3, yy.min() + .3, ('Eo=%.2f' % (1-score)).lstrip('0'), size=15, horizontalalignment='right')
            score = clf.score(X_train, y_train) 
            ax.text(xx.min() + .3, yy.min() + .3, ('Ei=%.2f' % (1-score)).lstrip('0'), size=15, horizontalalignment='left')
            
      else:
            ax.text(xx.max() - .3, yy.min() + .3, ('%.2f' % score).lstrip('0'), size=15, horizontalalignment='right')
  
      return score
    
  def plot_data_set(self, X, y, X_train, y_train,   i, title="Input data"):
      xx, yy = self.xx, self.yy
      width, height = self.width, self.height

      cm_bright, alpha = self.cm_bright, self.alpha
      ax = plt.subplot(width, height + 1, i)
       
      ax.set_title(title)

      # Plot the training points
      ax.scatter(X[:, 0], X[:, 1], c=y, cmap=cm_bright, alpha=alpha,
                edgecolors='k')
      # Plot the testing points
      ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright, alpha=alpha,
                edgecolors='k')
      ax.set_xlim(xx.min(), xx.max())
      ax.set_ylim(yy.min(), yy.max())
      ax.set_xticks(())
      ax.set_yticks(())
```

# Classification


## Definition

M-L program are system that builds (trains) a predictive model from input data. The system uses the learned model to make useful predictions from new (never-before-seen) data drawn from the same distribution as the one used to train the model.  

We will first look at M-L classifier, in our case, to solve binary classification problems. The task is to learn to distinguish  two discrete classes among available data and put a label on it. 

task that outputs one of two mutually exclusive classes.



## Components of learning

M-L algorithms are characteriszed by three main components: the input data $x$, a hypothesis set $H$ and the learning algorithm $A$. 




 # I. ML-algorithms

There are many ML-algorithms that exist for the task of binary classification, lets look at some of them.

## SVM

Support vector machine is a classic in machine learning for both regression and classification. It has been found to be particularly useful in applications where the number of attributes is much larger than the number of training objects. We are going to study two variant of the SVM. 






### Linear SVM

Before starting more advanced models we must first study the linear model. This linear model will be used as the reference model and is preffered as opposed to the perceptron because it gives us a unique solution, and it does not suffer from convergence problem. We first show it and use it later on to show that if any model cannot do better than the linear model, it is not intersting. 

**Maximum margin**

The SVM approach tries to computes the maximum margin hyperplane for a given set of data. The margin is the distance between the separation boundary and the closest samples. The goal is to find the optimal separating hyperplan, the one which separates
the two classes by maximizing the distance to the closest point from either class. These chosen point will be named support vectors.




| Maximum margin hyperplane |
|:---:|
|<img style="float: left; padding: 13px;" src="other/margin.png" width="180"> |

Above we find the margin, the large yellow band around the decision boundaries, in purple, and the chosen support vector, the dots with a circle around them.


 

**Optimisation problem**

the problem of finding such margin can be formulated as an optimisation problem:

Minimise  $$ {\frac  {1}{2}}||\mathbf{w}||^{2} \hspace{15mm}(1)$$   

by adjusting the weight $\mathbf{w}$ and the parameter $b$, subject to the constraints:

$$y_n(\mathbf{w}^Tx_n+b) \ge 0 \hspace{5mm} \forall n=1,...,N \hspace{15mm}(2)$$  




**Lagragian formulation**

What we have is called a constraint optimisation problem, where one must findg the values of a set of parameters - namely $\mathbf{w}$ and $b$ - that optimise an objective function but that also satisfy some constraints. This can be done using Lagrange multipliers, where the lagrangian is given by:

$$L(w,w_{0},\alpha )={\frac  {1}{2}}||\mathbf{w}||^{2}-\sum _{{n=1}}^{N}\alpha _{n}\left\{y_{n} \mathbf{w}^{T}x_{n}+b)-1\right\}\qquad \qquad \hspace{15mm}(3)$$



The optimum of the function (3) is attained when the derivative with respect to $\mathbf{w}$ and $b$ are equal to zero. By doing this we obtain the following equations:

$$\sum_{n=1}^N \alpha_n y_n x_n = \mathbf{w},\hspace{15mm}\sum _{{n=1}}^{N} \alpha _{n} y_n = 0 \hspace{15mm}(5)$$ 

 

By substituting these in the primal form we obtain the dual optimisation problem:

maximise for $\alpha:$ 

$$ L(\alpha)=
\sum _{i=1}^{N} \alpha _{i}-
{\frac {1}{2}} \sum _{i=1}^{M}\sum _{j=1}^{N} \alpha _{i} \alpha _{j} y_{i} y_{j}  x _{i} x _{j} 
\hspace{15mm}
(6)$$

Subject to the constraint:
$$\alpha _{n} \ge 0, \hspace{15mm}\sum _{{n=1}}^{N} \alpha _{n} y_n = 0 \hspace{15mm}(7)$$ 

As we can see in (6) this is a quadratic equation in $\alpha$, wich makes it quadratic programming problem (QP). Altough there is no analytical solution, this can be solve numericaly, with for example quadprog, a function from the MATLAB library.

**Support vector**

Now that the mathematical background has been set, we can define a couple of element,

Let, 
$$\{(x_1, y_1), ... , (x_n, y_n)\}\in \mathbb{R}^N\times\{1,-1\} \hspace{5mm}\forall n=1,...,N$$ 

be the training set with $y$ the labels and $N$ the dimension of the input vector $X$. 

The equation of the hyperplane $h(x)$ becomes :
$$h(x) = \sum _{{n=1}}^{N}\alpha _{n}\left\{y_{n} \mathbf{w}^{T}x_{n}+b)\right\}\hspace{15mm}(8)$$

One intersting result that comes with the last expression (6) and the KKT condition is:

$$\alpha _{n}\left\{y_{n} \mathbf{w}^{T}x_{n}+b)-1\right\}=0  \hspace{5mm}\forall n=1,...,N$$



from where two situation can happen: 

$$
\begin{equation}
 \begin{cases}
     y_{n} h(x_n)   = 1  \implies \alpha_n > 0 \hspace{5mm} x_n \text{  is on the boundary of the margin.  } \\
     y_{n} h(x_n) \ge 1  \implies \alpha_n = 0 \hspace{5mm}x_n\text{ is outside.  } 
  \end{cases}
\end{equation}
$$

The point $x_n$ for wich its lagragian multiplier $\alpha_n$ is greater than $0$ are called support vector. There are only a couple of them and these are the point used to delimitate the separation boundary.

**final hypothesis**

The task of classifying a new point $x_k$ from the training set or a new sample of data, is equivalent to, 

$$h(x_k) = sign\Big(\mathbf{w}^Tx_k + b\Big) = sign\Big(\sum _{{n=1}}^{K}\alpha _{n}\left\{y_{n} \mathbf{w}^{T}x_k+b \right\}\Big) \hspace{15mm}(9)$$

Notice that the algorithm only iterate on the $K$ support vector determined by the solver. The rest of the data point are not involved wich is very efficient in term of execution time.

**The C parameter**



| Fig.2 misclassification |
|:---:|
|<img src="other/slightly-non-separable.PNG" width="90" />|

In certain cases it is not possible to strictly classify every single point of the dataset in a linear way. As we see in (Fig. 2) some points might still lies outside of the decision boundary after the seperation. This is due to the presence of noise in the data, or because the two classes overlap.

To tackle this issue, we introduce the slack variable $\xi_n$ to (1.). The idea is to relax the constraints to allow some points close to the decision boudary to be misclassified. We now have:


 minimise 
 
$$ {\frac  {1}{2}}||\mathbf{w}||^{2}+ C \sum _{{n=1}}^{N} \xi_n  \hspace{15mm}(10)$$
 
by adjusting the variable $\xi_n$, $\mathbf{w}$ and $b$, subject to the constraints:

$$y_n(\mathbf{w}^Tx_n+b) \ge 1 - \xi_n \hspace{5mm}  \xi_n \ge 0 \hspace{5mm} \forall n=1,...,N \hspace{15mm}\hspace{15mm}(11)$$  

Where $\xi_n > 1$ happens when the misclassification of a point $x_n$ occur and $0 > \xi_n > 1$ otherwise. The constant C in the expression (11) can be seen as penalty.  This correpond to the width of the margin.

### Non linear SVM



We will now see a variant of the svm. The hypothesis is that in some case the data is not linearly separable. In some instance of a two dimensionaly classification problem, the data distribution can look like two concentric circle, a valley shape, etc. Therefore they cannot be seperated by a straight line anymore.

| Fig.4 non linear transform |
|:---:|
|<img src="other/n-linear-transform.png" width="380" />|
 

Consider the situation depicted in (Fig. 4) where the input vector $X \in R^N$ of the classification problem is not linearly separable. We define a new set of variables $Z \in R^M$ such that $z=x^{2}_{n1}+x^{2}_{n2}$ wich correspond to the representation of $x$ in fonction of its distance from the origin. Next, we map each $x \in X$ from the input space to its corresponding $z \in Z$ in what we call the feature space. Using $z$ instead of $x$ we can apply our learning algorithm explained before and it become possible to compute the maximum separation hyperplane. Therefore the procedure to classify a new point is:

transform $x$, classifiy $x$ get $y$, interpret $y$ back in the original $X$ space.

This time, we solve for z instead of x, the Lagragian formulation (6) becomes:

$$ L(\alpha)=
\sum _{i=1}^{N} \alpha _{i}-
{\frac {1}{2}} \sum _{i=1}^{M}\sum _{j=1}^{N} \alpha _{i} \alpha _{j} y_{i} y_{j}  z _{i} z _{j} 
\hspace{15mm}
(6)$$

**Kernel trick**

For any couple $(x_i,x_j) \in X$ the transformation $z_i z_j = K(x_i, x_j)$ is called a Kernel function.

A kernel function $K: X \times X \rightarrow \mathbb{R}^N$ is a symmetric, semi-defined positive function for all non-zero vectors $x_i,x_j \in X$.

Because in the constraint optimisation problem the variable $x_i, x_n, ..$ appear only as inner products $x_i x_i$ we can replace them by their correponding $K(x_i, x_j)$. The computation of the kernel is done in the input space,  no need to define a transformation for each and every value of $x$ in the $Z$ space, particulary if the dimension of $Z$ is large,  which is handy. Furthemore, the cost of calculating the inner product can be reduce to the evaluation of the Kernel in the input space the wich is less expensive. 

There is plenty of kernel function that exist, these are the three used in this notebook:

- linear: $K(x_n, x_m) = x^{T}_{n}x_m$.

- polynomial: $K(x_n, x_m)=(x_{n}^{T}\cdot x_{m}+1)^{\gamma}$.

- gaussian: ${K(x_n, x_m)=\exp(-\gamma \|x_n -x_m\|^{2})}$. 

**The gamma parameter**

The degree of the polynomial and the free parameter $\gamma$ in the gaussian kernel both appear as an additional parameter wich allow more flexibility. These are definied by the user and can be adjusted for a better decision boundary. 

## KNN

The nearest neighbor algorithm is simple but also very efficient. The principle is that data points are classified according the K-closest neighbours around it.  

**Learning Algorithm**

The training examples are vectors in a multidimensional feature space, each with a class label. The training phase of the algorithm consists only of storing the feature vectors and class labels of the training samples. In the classification phase, k is a user-defined constant, and an unlabeled vector is classified by assigning the label which is most frequent among the k training samples nearest to that query point.



There is no training phase so to speak, the algorithm consists only of storing the feature vectors and class labels of the training samples. The entire algorithm is specified by how one computes the final hypothesis $g(x)$ on a test input $x$. 

To classify the test point $x$, find the nearest point to $x$ in the data set (the nearest neighbor), and use the classification of this nearest neighbor. Formally speaking, reorder the data according to distance from $x$. The metric used for the distance is the usual euclidian distance. We write $(x_{[n]}(x), y_{[n]}(x))$ for the $nth$ such reordered data point with respect to x. We will drop the
dependence on x and simply write $(x_{[n]}, y{[n]})$ when the context is clear. So

$d(x, x[1]) ≤ d(x, x[2]) ≤ · · · ≤ d(x, x[N])$

The final hypothesis is

$g(x) = y_{[1]}(x)$

**Parameter Selection**



The best choice of k depends upon the data; generally, larger values of k reduces effect of the noise on the classification,[5] but make boundaries between classes less distinct. A good k can be selected by various heuristic techniques (see hyperparameter optimization). The special case where the class is predicted to be the class of the closest training sample (i.e. when k = 1) is called the nearest neighbor algorithm.

The accuracy of the k-NN algorithm can be severely degraded by the presence of noisy or irrelevant features, or if the feature scales are not consistent with their importance. Much research effort has been put into selecting or scaling features to improve classification. A particularly popular[citation needed] approach is the use of evolutionary algorithms to optimize feature scaling.[6] Another popular approach is to scale features by the mutual information of the training data with the training classes.[citation needed]

In binary (two class) classification problems, it is helpful to choose k to be an odd number as this avoids tied votes. One popular way of choosing the empirically optimal k in this setting is via bootstrap method.[7]

A drawback of the basic "majority voting" classification occurs when the class distribution is skewed. That is, examples of a
more frequent class tend to dominate the prediction of the new example, because they tend to be common among the k nearest neighbors due to their large number.[4] One way to overcome this problem is to weight the classification, taking into account the distance from the test point to each of its k nearest neighbors. The class (or value, in regression problems) of each of the k nearest points is multiplied by a weight proportional to the inverse of the distance from that point to the test point. Another way to overcome skew is by abstraction in data representation. For example, in a self-organizing map (SOM), each node is a representative (a center) of a cluster of similar points, regardless of their density in the original training data. K-NN can then be applied to the SOM.

# Performance Analysis

## Testing phase

Lets first start by testing all the models on the training dataset with defaut settings. Starting with the algorithm in use.

```python
titles = ["SVM linear", "SVM poly 2", "SVM poly 3", "SVM RBF", "KNN"]

classifiers = [ 
    SVC(kernel = 'linear' ),
    SVC(kernel = 'poly', degree=2 ),
    SVC(kernel = 'poly', degree=3 ),
    SVC(kernel = 'rbf'),
    KNeighborsClassifier(n_neighbors=3)]
```

Next is the training phase. We also plot the decision boundary using a helper function.

```python

plotter = Plotter( len(datasets), len(classifiers))
i=1

scores = {t:[] for t in titles}
    
for ds_cnt, ds in enumerate(datasets):
    # preprocess dataset, split into training and test part
    X, y = ds
    X = StandardScaler().fit_transform(X)
    X_train, X_test, y_train, y_test = \
        train_test_split(X, y, test_size=.33)

    plotter.setup(X) 
    plotter.plot_data_set(X, y, X_train,y_train,i,dataset_titles[ds_cnt])
    i += 1

    # iterate over classifiers
    for title, clf in zip(titles, classifiers):
        # training phase
        clf.fit(X_train, y_train) 
        score = plotter.plot_result(X_train, y_train, X_test, y_test,ds_cnt, clf, i, title )
        scores[title].append(score)
        i += 1

plt.plot()
```

```python
t1 = [[key, sum(val)/len(val)]for key, val in scores.items()]

sorted_t1 = sorted(t1,key=lambda l:l[1], reverse=True)

print(tabulate( sorted_t1  , headers=['model', 'mean score'])) 
```

We can assess that they all perform better than the linear model, beside the 3d polynomial kernel. The rbf gets the upper hands in the initial phase.

Next we will see how to get improve the performance of the models.

## Noisy Dataset

The dataset contains noise. There might be some mislabeled colors that do not correpond to their true x-values, or corruption in the dataset occuring during manipulation, etc. The label assigned to the feature is not 100 percent representative, therefore we consider the dataset as a probabilistic distribution and we need to take precaution with the noise that comes with it.



In practice, an effective way to reduce the influence of noise is to use cross validation. Basicly, we split the dataset in many folds, 10 for regular dataset, we then apply the learning process on each individual fold. At the same time, a grid search is performed during the process and the overall best parameter is selected.

As we saw above, for the svm, the lagrangian formulation involved many parameter.

The C parameter tells the SVM optimization how much you want to avoid misclassifying each training example. For large values of C, the optimization will choose a smaller-margin hyperplane if that hyperplane does a better job of getting all the training points classified correctly. Conversely, a very small value of C will cause the optimizer to look for a larger-margin separating hyperplane, even if that hyperplane misclassifies more points.

There is also Gamma.

**GridSearch**

Lets do a quick search to find the most optimal parameter

```python
 
# Define a list of parameters for the models
params = {'C': [0.1,1, 10, 100], 'gamma': [1,0.1,0.01,0.001], 'kernel': ['rbf', 'poly', 'linear']}
scores, best_params = [], []

for ds_cnt, ds in enumerate(datasets):
    X, y = ds
    X = StandardScaler().fit_transform(X)
    X_train, X_test, y_train, y_test = \
        train_test_split(X, y, test_size=.4, random_state=42)
    # We can build Grid Search model using the above parameters. 
    # cv=5 means cross validation with 5 folds
    grid_search = GridSearchCV(SVC(), params, cv=3, n_jobs=-1 ,scoring='accuracy' ) 
    grid_search.fit(X_train, y_train)
    scores.append(grid_search.score(X_test, y_test))
    best_params.append(grid_search.best_params_) 
    

tabs = [elem for elem in zip(dataset_titles, scores, best_params)]

print(tabulate(tabs, headers=['dataset', 'score', 'optimal params'])) 
```

Now we found the optimal parameter, let us compare with the default parameter.

```python

 
titles = ["linear", "optimized"]

default = SVC(kernel = 'linear' )

classifiers, optimal = [], None

for params in best_params: 
    optimal = SVC(kernel=params["kernel"], C=params["C"], gamma=params["gamma"])
    classifiers.append(optimal)

plotter = Plotter( len(datasets), len(titles)) 
i=1 
scores = {t:[] for t in titles}

for ds_cnt, ds in enumerate(datasets):
    # preprocess dataset, split into training and test part
    X, y = ds
    X = StandardScaler().fit_transform(X)
    X_train, X_test, y_train, y_test = \
        train_test_split(X, y, test_size=.33)

    plotter.setup(X)
    plotter.plot_data_set(X, y, X_train, y_train,   i, dataset_titles[ds_cnt])
    i += 1
        
    # iterate over classifiers
    for title in titles:
        clf = None
        if title == "linear":
            clf = default
        else: 
            clf = classifiers[ds_cnt]
            
        clf.fit(X_train, y_train)
        score = plotter.plot_result(X_train, y_train, X_test, y_test, ds_cnt, clf, i, title )
        scores[title].append(score)
        i += 1
        
plt.plot()

```

```python
t1 = [[key, sum(val)/len(val)]for key, val in scores.items()]
    
print(tabulate( t1  , headers=['model', 'score'])) 
```

As we can see, the use of cross validation to tune the parameter helped use increase the overall accuracy.

## Evaluating the estimator

In the process of learning there is this error during the learning problem associate with the fact that we can not learn 100 percent correctly. The final hypothesis tries to approximate the target function as close as possible, plus minus an error.

One of the fondamentals requirement for any learning algorithm is it should be complex enough so we can fit the data and obtain low $E_{in}$; however, it should be reliable enough that a low $E_{in}$ implies a low $E_{out}$. This is the ideal situation but it does not always happens to be that way.
 

<img style="float: left; padding: 13px;" src="other/overfitting.png" width="280">

The figure on the left tells us two case scenario. 

The first happens on the right side. As the complexity of the model increase, so does the chance of approximating $f$. But what we observe is that the model does poorly on out-of-sample data even though the in-sample error keep on decreasingµ. This means the model is very performant in the traning phase, to the point where it memorize the training set. It then becomes difficult for the model to generalize to new examples. This is called overfitting.

Secondly, the less complex the model the better the chance of generalizing out of sample is, also the learning gap between $E_{in}$ and $E_{out}$.

The balance is somewhere between the two, at the inflexion point of $E_{out}$.

**Overfitting**

The first scenario is illustrated here. Two models are trained, one is a SVM with default parameter, the other one is a variant in wich its learning capacity are pushed to the limit for the purpose of explanation. 

```python
titles = ["default", "overfitting"]

classifiers = [ 
    SVC(kernel="rbf"    ),
    SVC(kernel = "rbf",   C=1000, gamma=10 ) ]
 
i=1 
scores = {t:[] for t in titles}  
plotter = Plotter( 4, len(classifiers))

for ds_cnt, ds in enumerate([datasets[0]]):
    # preprocess dataset, split into training and test part
    X, y = ds
    X = StandardScaler().fit_transform(X)
    X_train, X_test, y_train, y_test = \
        train_test_split(X, y, test_size=.33)

    plotter.setup(X) 
    plotter.plot_data_set(X, y, X_train,y_train,i,dataset_titles[ds_cnt])
    i += 1

    # iterate over classifiers
    for title, clf in zip(titles, classifiers):

        clf.fit(X_train, y_train) 
        score = plotter.plot_result(X_train, y_train, X_test, y_test,ds_cnt, clf, i, title, inSample=True )
        scores[title].append(score)
        i += 1

plt.plot()
```

As we can on the right side, the model is doing great in the training phase, but on the testing set it performed poorly compared to the default model. This shows that the model is overfitting.

**Learning curve**

As we saw above, to prevent overfitting the learning algorithm should stop early. We set the stopping point in such a way that when the performance has stopped improving on the validation set, the model training stops. 

Some algorithm do implement an iterative mechanism that let you compute the model per iteration. For example, in XGBoost, as you train more and more trees, you will overfit your training dataset, this also exist for SVM. Early stopping enables you to specify a validation dataset and the number of iterations after which the algorithm should stop if the score on your validation dataset didn’t increase.

The code below does three thing:
 
- Split the training data into a training set and a validation set.
- Train the model and once in a while perform a validation.
- Plot the error on the validation set and the training set for comparaison.

Thankfully most of the heavy weight is done by the librairies. All thats left is to specify the start, the stop and the increment size according to the training set. 

```python
def learning_curve_model(estimator, title, X, y, cv=None,
                        n_jobs=7, train_sizes=np.linspace(0.1, 1.0, 5)):
    # estimator  : the model
    # X,y        : the label and data
    # cv         : is the cross validation method to evaluate the score
    # n_jobs     : parameter for multithreading capabilities
    # train_size : range of the learning curve (start, stop, step)
    
    plt.figure()
    plt.title(title)
    plt.xlabel("Training examples")
    plt.ylabel("Score")


    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    
    train_scores_mean = 1- np.mean(train_scores, axis=1)
    train_scores_std  =  np.std(train_scores, axis=1)
    test_scores_mean  = 1- np.mean(test_scores, axis=1)
    test_scores_std   =  np.std(test_scores, axis=1)
    
    plt.grid() 
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std, train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std, test_scores_mean + test_scores_std, alpha=0.1, 
                     color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",label="E_in")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",label="E_out") 
    plt.legend(loc="best")
    
 
 
```

```python

X,y = datasets[0]
estimator = optimal 

title = "Learning Curves of the optimized model"  

cv = ShuffleSplit(n_splits=5, test_size=0.4 )  
learning_curve_model(estimator, title, X, y, cv=cv )
```

From 85 training data points onward, the validation score stays roughly the same. This means that adding more training data points won't lead to significantly better models. The gap is low also and the two curve do not meet. The model is well balanced. It doesnt show too much biais or variance.

## Digit recognizer

The task of recognizing digit is fairly obvious to a human, but even with manual observation there is still a bit of error. We will show how to train a learning algorithm to recognize a number from 0 to 9 encoded in an pixel image.


The data files train.csv and test.csv contain gray-scale images of hand-drawn digits, from zero through nine. Each image is 28 pixels in height and 28 pixels in width, for a total of 784 pixels in total. Each pixel has a single pixel-value associated with it, indicating the lightness or darkness of that pixel, with higher numbers meaning darker. This pixel-value is an integer between 0 and 255, inclusive.

The training data set, (train.csv), has 785 columns. The first column, called "label", is the digit that was drawn by the user. The rest of the columns contain the pixel-values of the associated image.

```python
nRowsRead = None # specify 'None' if want to read whole file
df_train = pd.read_csv('dataset/train.csv', delimiter=',', nrows = nRowsRead)
df_train.dataframeName = 'Training set'
nRow, nCol = df_train.shape
print(f'There are {nRow} rows and {nCol} columns')
```

```python
df_train.head(5)
 
```

**Preprocessing**

We will have to keep only 2 features as we need to see the boundary decision for testing purpose. From the input pixel, we will compute two basic value: intensity and symmetry. Intensity is the amount of black pixel covering the image. Symmetry on the other hand is a mesure wich tells us of much the pixels overlap along a given axis, say y-axis. for example the digit 8 might have a bigger overlap than the number 5.

```python
# compute the average pixels
df1 = df_train.mean(axis=1)
# compute the simmetry along the y-axis
h = 24
cols = nCol
iter = int(cols/h)
 
df2 = df_train.loc[:,['label']]
df2["sum1"] ,df2["sum2"] = 0, 0

 
for i in range(iter):
    off, off2 = i * h , (i+1)*h
    off1 = int((off + off2)/2)
 
    off, off1, off2  = off+1, off1+1, off2+1
 
    df2["sum1"] += df_train.iloc[:,off:off1].sum(axis=1)
    df2["sum2"] += df_train.iloc[:,off1:off2].sum(axis=1)

# substract the left side with the right symmetric side along the y-axis
df2["sub"] = df2["sum1"] - df2["sum2"] 
# make a dataframe from the new computed value
df = pd.DataFrame( { "label": df_train["label"], "intensity": df1, "symmetry": df2["sub"]})

df.describe()
 
```

```python
df.head(5)
```

```python
df.plot.scatter(x="intensity", y="symmetry", c="label", alpha=0.7, 
                      colormap='viridis')
```
